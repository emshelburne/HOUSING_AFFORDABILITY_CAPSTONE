## Helper functions

```python
# Define function to examine dataframes

def examine_df(name,df):
    """
    Check basic info about a dataframe df
    """
    
    print(f"\n\nNumber of records in the {name} is: {len(df)}\n")
    print(f"The columns in the {name} are: {df.columns}\n")
    print(f"\n Other info about {name}:")
    display(df.info())
    print(f"\n\nSample of records in the {name}:")
    display(df.head(5))
```

```python
# Load the zones and nbhds shapefiles
zones_gdf = gpd.read_file("DATA/GEOGRAPHIC/ZONE_2023/Zone_2023/Zone_2023.shp")
nbhds_gdf = gpd.read_file("DATA/GEOGRAPHIC/NBHD_2023/Nbhd_2023/Nbhd_2023.shp")
```

## Clean and Preprocess Zones Dataframe

```python
# Drop unnecessary columns
columns_to_keep = ['METZONE_UI', 'METCODE', 'ZONECODE', 'ZONENAME_E', 'geometry']
zones_gdf = zones_gdf[columns_to_keep]
```

```python
print(len(zones_gdf))
zones_gdf['METCODE'].unique()
```

```python
# We know from CMHC that Vancouver has 25 survey zones so we use this to determine which METCODE corresponds to Vancouver
zone_counts = zones_gdf.groupby('METCODE').size().reset_index(name='zone_count')
likely_vancouver = zone_counts[(zone_counts['zone_count'] >= 20) & (zone_counts['zone_count'] <= 30)]
display(likely_vancouver)
# We find 2410 = Vancouver
```

```python
# Confirm we have located the correct METCODE for Vancouver
vancouver_zones = zones_gdf[zones_gdf['METCODE'] == '2410']
vancouver_zones[['ZONECODE', 'ZONENAME_E']].sort_values('ZONECODE')
```

```python
# Define new Vancouver specific dataframe for survey zones
van_zones_df = zones_gdf[zones_gdf['METCODE'] == '2410'].copy()
van_zones_df.drop(columns = ['METCODE'])
```

```python
# Examine new dataframe
examine_df('Vancouver Zones Dataframe',van_zones_df)
```

## Clean and Preprocess Neighborhood Dataframe

```python
# Drop unnecessary columns
display(nbhds_gdf.columns)
columns_to_keep = ['METNBHD_UI', 'METCODE', 'NBHDCODE','NBHDNAME_E','ZONECODE', 'geometry']
nbhds_gdf = nbhds_gdf[columns_to_keep]
```

```python
# Define new Vancouver specific dataframe for neighborhoods
van_nbhds_df = nbhds_gdf[nbhds_gdf['METCODE'] == '2410'].copy()
van_nbhds_df.drop(columns = ['METCODE'])
```

```python
# Examine new dataframe
examine_df('Vancouver Zones Dataframe',van_nbhds_df)
```
